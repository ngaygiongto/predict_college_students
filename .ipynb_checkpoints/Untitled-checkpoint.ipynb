```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('Sinvien.xlsx', index_col = 'STT')
```

```python
data.head()
```

```python
data.info()
```

## Kiểm tra giá trị nan

```python
data.isna().sum()
```

```python
data.describe()
```

## Replace 

```python
data['XL2'].value_counts()
```

```python
data['XL2'].replace(['Antoan', 'Nguyco'], [0,1], inplace = True)
```

## Hiển thị sự tương quan dữ liệu

```python
data_corr = data.corr()
```

```python
data_corr
```

```python
plt.figure(figsize=(15,13))
sns.heatmap(data_corr, annot=True, cmap='viridis_r')
```

```python
X = data.iloc[:,:-1]
```

```python
Y = data['XL2']
```

## Train Test Split


```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size = 0.3, random_state=42)
```

## SVM model

```python
from sklearn.svm import SVC
```

```python
model_svc = SVC()
```

```python
model_svc.fit(x_train,y_train)
```

## Prediction and Evaluation

```python
y_pred_svc = model_svc.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, y_pred_svc))
```

```python
confusion_matrix = confusion_matrix(y_test, y_pred_svc)
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Dự đoán')
plt.ylabel('Thực tế')
plt.title('Ma trận nhầm lẫn')
plt.show()
```

## Gridsearch

```python
from sklearn.model_selection import GridSearchCV
```

```python
help(SVC)
```

```python
parameters = {
    'C': [0.1,1,10,100,1000],                     # C cang lon bias giam, phuong sai tang
#     'kernel':['poly','rbf', 'sigmoid'],
#     'degree':[3, 4, 5],
    'gamma':[0.0001, 0.001, 0.01, 0.1, 1],        # gamma cang lon do bias cang lon, phuong sai giam
}
```

```python
grid = GridSearchCV(SVC(), parameters, cv= 5, verbose=2, refit=True)
```

```python
grid.fit(x_train, y_train.values.ravel())
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
predictions = grid.predict(x_test)
```

```python
print(classification_report(y_test, predictions))
```
